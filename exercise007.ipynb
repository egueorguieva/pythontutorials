```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data007 = pd.read_csv("/Users/emmagueorguieva/git/pythontutorials/datasets/007ExerciseFile.csv")
```

```python
display(data007)
```

First, I'll make a histogram for all four strains:

```python
sns.displot(data007, x="Mutation_rate", hue="Strain", kind="hist", alpha=0.3)
```

That... looks kind of terrible. Next, I'll do a KDE plot to see if I can visualize the data a little better.

```python
sns.displot(data007, x="Mutation_rate", hue="Strain", kind="kde", fill=True, alpha=0.2, bw_adjust = 0.6)
```

That looks a lot better. I'll make a violin plot next.

```python
sns.violinplot(data=data007, x="Mutation_rate", y="Strain")
```

It's interesting to look at, but I don't really like this one this one to showcase the differences in the data. I'll try a boxplot next.

```python
sns.boxplot(data=data007, x="Mutation_rate", y="Strain")
```

This is pretty nice! In this circumstance, I think it's easier to understand than the violin plot I made. Let me add a strip chart.

```python
sns.boxplot(data=data007, x="Mutation_rate", y="Strain")
sns.stripplot(data=data007, x="Mutation_rate", y="Strain", alpha = 0.42)
```

I think this one is also really pretty to look at. In the end, I think that the boxplot with overlaid strip chart is the best graph to visually show that the strains do not mutate at the same rate. You can clearly see there is little to very little overlap of the error bars which indicates a significant difference. (With the exception of st_delta and st_gamma, which are pointing to no significant difference. However, there is *enough* difference in the other strains for this plot to still be relevant.) Now, let me clean this chart up and make it look pretty for Dr. Fauci.

```python
plot = sns.boxplot(data=data007, x="Mutation_rate", y="Strain")
sns.stripplot(data=data007, x="Mutation_rate", y="Strain", alpha = 0.42)
plot.set(xlabel = "Mutation Rates")
plot.set_yticklabels(["Alpha", "Beta", "Gamma", "Delta"])
```
